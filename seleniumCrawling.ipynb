1. 키노라이츠 넷플릭스의 영화 키값 가져오기

```python
url = "https://m.kinolights.com/discover/explore"

def get_chrome_driver():
    # 1. 크롬 옵션 세팅
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--disable-gpu')
    # chrome_options.add_argument("--headless")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-logging")
    chrome_options.add_argument("--disable-gpu")

    # 2. driver 생성하기
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()), # 가장 많이 바뀐 부분
        options=chrome_options    
    )
    
    return driver

driver = get_chrome_driver()
driver.get(url)

# 넷플릭스 컨텐츠 선택 후 영화 버튼 선택
netflix_button = driver.find_element_by_xpath('//*[@id="commentArea"]/div[1]/div[1]/div/div/button[1]')
movie_button = driver.find_element_by_xpath('//*[@id="commentArea"]/div[2]/div[1]/button[2]')
netflix_button.click()
movie_button.click()



SCROLL_PAUSE_SEC = 2
# 스크롤 높이 가져옴
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # 끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # 1초 대기
    time.sleep(SCROLL_PAUSE_SEC)
    # 스크롤 다운 후 스크롤 높이 다시 가져옴
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height


# xpath로 key-data path 선택
key_data = driver.find_elements(
    By.XPATH,
    '//*[@id="listArea"]/div[3]/div/div')
len(key_data)


# xpath로 접근하여 속성값에서 'data-key' 값을 추출하여 리스트로 저장
netflixData_list = []
for i in range(0,len(key_data)):
    name = key_data[i].get_attribute('data-key')
    netflixData_list.append(name)
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
import urllib.parse as req
import time , pymysql
import warnings
import os
warnings.filterwarnings("ignore")
# selenium hide logs
import logging
logging.disable(logging.CRITICAL)
url = "https://m.kinolights.com/title/new"

def get_chrome_driver():
    # 1. 크롬 옵션 세팅
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-logging")
    os.environ['WDM_LOG_LEVEL'] = "0" # 불필요한 로그를 출력하지 않도록 함.
    # 2. driver 생성하기
    driver = webdriver.Chrome(options=chrome_options, service=Service(ChromeDriverManager().install()))
    return driver


driver = get_chrome_driver()
driver.get(url)

# 크롤링하고자 하는 OTT 클릭
netflix_button = driver.find_element_by_xpath('//*[@class="kino-icon kino-icon--netflix"]')
wavve_button = driver.find_element_by_xpath('//*[@class="kino-icon kino-icon--wavve"]')
watcah_button = driver.find_element_by_xpath('//*[@class="kino-icon kino-icon--watcha-play"]')
diseny_button = driver.find_element_by_xpath('//*[@class="kino-icon kino-icon--disney"]')

netflix_button.click()
time.sleep(0.5)
wavve_button.click()
time.sleep(0.5)
watcah_button.click()
time.sleep(0.5)
diseny_button.click()
time.sleep(0.5)


key_data = driver.find_elements(
    By.XPATH,
    '//*[@id="listArea"]/section[*]/div/div/div/div/div/div')
len(key_data)

# xpath로 접근하여 속성값에서 'data-key' 값을 추출하여 리스트로 저장
new_movieList = []
for i in range(0,len(key_data)):
    name = key_data[i].get_attribute('data-key')
    new_movieList.append(name)
new_movieList



def get_driver(target_url):
    os.environ['WDM_LOG_LEVEL'] = "0"
    time.sleep(2)
    driver = get_chrome_driver()
    driver.get(target_url)
    wait=WebDriverWait(driver, 3)
    return driver

def extract_data(driver):
    title_kr = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div[5]/div/h3').text.strip()
    genre = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div[5]/div[1]/p/span[1]').text.replace("·","")
    try:
        release_date = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div[5]/div[1]/p/span[2]').text.replace("·","")
    except:
        release_date = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div[5]/div[1]/p/span').text.replace("·","")
    
    rotten_rate = driver.find_element_by_class_name('rotten-wrap').text.strip()
    imdb_rate = driver.find_element_by_class_name('imdb-wrap').text.strip()
    star_rate = driver.find_element_by_class_name('movie-star-wrap').text.strip()
    synopsis = driver.find_element_by_class_name('synopsis').text.strip()
    director = driver.find_element_by_xpath('//*[@id="synopsis"]/article/div/div[1]/div[1]').text.replace("·","")
    runtime = driver.find_element_by_xpath('//*[@id="contents"]/div[4]/ul/li[1]').text.replace("#","")
    

    locates = "div.name"
    test_actor = driver.find_elements_by_css_selector(locates)
    test_actor = test_actor[1:]
    test_actor
    actors_list = []
    for test_actors in test_actor:
        test_actors = test_actors.text
        actors_list.append(test_actors)
    actors_list = ",".join(actors_list)

    poster_link = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div[5]/div[2]/img').get_attribute('src')

    # OTT 연결 링크
    movie_links = driver.find_elements_by_xpath('//*[@id="contents"]/section[1]/div/div/ul/li[*]')
    href = driver.find_elements_by_xpath('//*[@id="contents"]/section[1]/div/div/ul/li[*]/div/a')

    lst_links =[]
    for i in href:
        linkUrl = i.get_attribute('href')
        lst_links.append(linkUrl)    
    lst_links = ",".join(lst_links)

    # 키노라이츠 지금 여기서 감상 서비스 이름 추출

    allOttNames = driver.find_elements_by_xpath('//*[@id="contents"]/section[1]/div/div/ul/li[*]/div/a/div/span[2]')
    ottnames = []
    for name in allOttNames:
        ottName = name.text
        ottnames.append(ottName)
    ottnames = ",".join(ottnames)


    title_list.append(title_kr)
    release_date_list.append(release_date)
    rotten_rate_list.append(rotten_rate)
    imdb_rate_list.append(imdb_rate)
    star_rate_list.append(star_rate)
    genre_list.append(genre)
    runtime_list.append(runtime)
    director_list.append(director)
    synopsis_list.append(synopsis)
    ac_list.append(actors_list)  
    ott_name.append(ottnames)
    ott_link.append(lst_links)
    poster_list.append(poster_link)


    print('한글:',title_kr)
    # print("장르:",genre, "개봉일:",release_date,"상영시간:",runtime)    
    # print("평점",rotten_rate,"," ,imdb_rate,",", star_rate)
    # print("시놉시스:",synopsis)
    # print("감독:",director)
    # print("출연진:",actors_list)
    # print("여기서볼래:",ottnames)
    # print("여기서볼래링크:", lst_links)

for i in new_movieList:
    target_url = "https://m.kinolights.com/title/{}".format(i)
    crawling = get_driver(target_url)
    extract_data(crawling)

print("스크래핑이 완료되었습니다. selenium을 종료 합니다.")
# crawling.quit()
```
